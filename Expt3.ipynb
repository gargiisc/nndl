```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def binaryStep(x):
    return np.heaviside(x, 1)

x = np.linspace(-10, 10)
plt.plot(x, binaryStep(x))
plt.axis('tight')
plt.title('Activation Function: binaryStep')
plt.show()
```

```python
def linear(x):
    return x

x = np.linspace(-10, 10)
plt.plot(x, linear(x))
plt.title('Activation Function: Linear')
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.title('Activation Function: Sigmoid')
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.title('Activation Function: Tanh')
plt.show()
```

```python
def RELU(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)
    return x1

x = np.linspace(-10, 10)
plt.plot(x, RELU(x))
plt.title('Activation Function: RELU')
plt.show()
```

```python
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x), axis=0)

x = np.linspace(-10, 10)
plt.plot(x, softmax(x))
plt.title('Activation Function: Softmax')
plt.show()
```
